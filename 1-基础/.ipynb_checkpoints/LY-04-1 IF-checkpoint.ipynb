# 双向循环
- if...else...语句

  if 条件表达式:
      语句1
      语句2
      ...
  else:
      语句1
      语句2
      .....
- 双向分支有两个分支，当程序执行到if ... else...语句的时候，一定会执行if或else中的一个，也仅执行一个
- 缩进问题，if和else一个层级，其余语句一个层级

```python
# input()的作用：
#1、先输出括号内的字符串。
#2、接受用户输入的内容，并返回给程序。
#3、返回的值一定是个字符串类型
gender = input('请输入性别:')
print ('你输入的性别是{0}'.format(gender))
```

```python
if  gender == "nan":
    print("来，我们纪念一下今天吧，代码敲十遍")
else:
    print("发糖喽发糖喽")
    print("你是女生，特殊照顾喽")
        
print("开始上课喽")
```

# 多路分支
- 很多分支的情况，简称多路分支

  if  条件表达式：
      语句1
      ....
  elif 条件表达式：
      语句1
      ...
  elif 条件表达式：
      语句1
      ...
  .....
  else:
      语句1
      ....
      
- elif 可以由很多

- else可选
- 多路分支只会选一个执行

```python
# score  存放学生成绩
# 注意input的返回值类型
score = input("请输入学生成绩：")
# 需要把str转换成int
score = int(score)

if score>=90:
    print("A")
elif score>= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >=60 :
    print("D")
else:
    print("起开，我没你这傻学生")
```

# 循环语句
- 重复执行某些固定动作或者处理基本固定的事物
- 分类
  for循环
  while循环 # for循环
- for循环

  for 变量 in 序列：
      语句1
      语句2
      ...

```python
list1 = ['al','sk','ij','ok']
for i in list1:
    print(i)
```

```python
list1 = ['al','sk','ij','ok','op','er']
for i in list1:
    print(i)
    if i == 'ok':
        print('找到{0}啦!!!!'.format(i))
    else:
        print("继续找")
        
```

# range介绍
- 生成一个数字序列
- 范围可以设定

```python
# 注意，一般在python中，如果由表示数字范围的两个数，一般是包含左边数字不包含右边数字
# randint是特例，他左右都包含
# range函数在python2 和 python3中有严重区别

for i in range(1,11):
    print(i)
```

# for-else语句
- 当for循环结束的时候，会执行else
- else语句是可选语句

```python
# for-else语句
# 打印列表中的同学，
# 如果没有在列表中，或者列表结束了，我们需要打印提示语句，表示不在爱了

# 打印学生列表姓名
# 如果是jingjign，那肯定是我的最爱呀，
# 如果是别的学生，那要冷酷的拒绝他
for name in  ['zhangsan', 'lisi', 'wangwu','jingjing']:
    print(name)
    if name == "jingjing":
        print("我的最爱{0}出现了".format(name))
    else:
        print("同学我们不约，不约，同学请自重")
else:
    print("别的都不是我的学生，不会在爱了")
```

# break  contineu  pass
- break：无条件结束整个循环，简称循环猝死
- contineu：无条件结束本次循环，继续进入下一轮循环
- pass：表示略过，通常用于占位

```python
# for 循环中的变量表示，一般用i，k，m，n，或者indx，idx，item之类
for i in range(1,11):
    if i == 6:
        print('终于找到{0}啦！'.format(i))
        break
    else:
        print(i)

# 在python中，如果循环变量名称不重要，可以用下划线（_)代替        
for _ in range(10):
    print('* ',end='')
```

```python
# 找偶数
for i in range(1,11):
    if i%2 ==1:
        continue
    else:
        print('偶数为：{0}'.format(i))
        
print('='*20)

# 找偶数2
for i in range(1,11):
    if i%2 == 0:
        print('偶数为：{0}'.format(i))
        
print('='*20)

# 本案例完整显示continue的用法和作用
for i in range(1,11):
    if i % 2 == 0:
        continue

        
    print('奇数为：{0}'.format(i))
```

```python
# pass例子，一般用于占位
# pass没有跳过功能

for i in range(1,3):
    pass
    print("wo zai zheli")
```

# while循环
- 在条件成立之前保持循环
- 不知道循环次数
- while语法
      while 条件表达式:
          语句块
      
      while 条件表达式：
          语句块1
      else:
          语句块2

```python
# 年利率6.7%，本利翻滚，几年后翻倍。

b = 100000
year = 0

while b<200000:
    b *= (1 + 0.067)
    year += 1
    print('第{0}年，目前额度：{1}元'.format(year,b))
else:
    print('hahaha')
```
