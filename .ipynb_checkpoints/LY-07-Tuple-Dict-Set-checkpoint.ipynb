```python
# count() 计算某个元素在元组中出现的次数xxxxxxxxxxxxxxxxxxxxxxx
tuple1 = (3,2,1,4,3,6)
print(tuple1.count(3))
```

# Dict 字典

```python
help(dict)
```

```python
# clear()清除整个字典，不是删除，返回None
dict1 = {'a':1,'b':2,'c':3}
dict1.clear()
print(dict1)
```

```python
# copy()复制字典
dict1 = {'a':1,'b':2,'c':3}
dict2 = dict1.copy()
print(dict2)
```

```python
# formkeys()按照指定的列表为键创建字典，值都是一样的。
list1 = ['a','b','c']
dict1 = {}.fromkeys(list1)
dict2 = {}.fromkeys(list1,3)
print('\n',dict1,'\n',dict2)
```

```python
# get()根据指定的键获取值，找不到的键如果给了值则返回值，如果没给值则返回None
dict1 = {'a':1,'b':2,'c':3}
print(dict1.get('a'))
print(dict1.get('d'))
print(dict1.get('d',5))
```

```python
# setdefault() 在字典里先查找，并更新或添加一个元素  
# 类似get(k,d) ，差别是 如果没找到，则添加到末尾，并返回默认给定值
dict1 = {'d':4,'a':1,'e':2,'y':6}
print(dict1.setdefault('d'))#找到了，返回默认给定值
print(dict1.setdefault('d',5))#找到了，忽略给定值给定值
print(dict1.setdefault('xx',5))#没找到，添加到末尾，并返回默认给定值
print(dict1)
```

```python
help(dict.setdefault) 
```

```python
# item()将字典变成类似于元组的形式方便遍历
dict1 = {'a':1,'b':2,'c':3}
#用k,v两个变量遍历每一个数据
for k,v in dict1.items():
    print(k,v)

#用i遍历每一组元组
for i in dict1.items():
    print(i)

print(dict1.items())
```

```python
help(dict.items)
```

```python
# pop()移除字典中指定元素，返回键所对应的值，如果键不存在，则返回默认值，如果键找不到，没设默认值，就会报错
dict1 = {'a':1,'b':2,'c':3}
print(dict1.pop('a'))
print(dict1)
print(dict1.pop('d',4))
print(dict1)
print(dict1.pop('d'))
print(dict1)
```

```python
# popitem() 移除末尾的键值对，并返回移除的键值对
dict1 = {'a':1,'b':2,'c':3}
print(dict1.popitem())
```

```python
help(dict.popitem) 
```

```python
# update() 用新的字典键值对，更新修改原有字典中的值，原有字典地址不变
dict1 = {'d':4,'a':1,'b':2,'c':3}
dict1.update({'a':88,'b':99,'c':100})
print(dict1)
print(id(dict1))

dict1.update({'a':88,'b':99,'c':100,'d':101})
print(dict1)
print(id(dict1))

dict1.update({'x':112,'a':88,'b':99,'c':100,'d':101,'f':102})
print(dict1)
print(id(dict1))
```

# Set 集合

```python
help(set)
```

```python
a = set()
print(a)
list1 = [1,2,3,4]
a = set(list1)
print(a)
```

```python
# add()向集合中添加元素
set1 = {5,1,2,3,4,'b','u'}
set1.add(6)
print(set1)
```

```python
# clear() 清空
# copy() 复制
# pop() 随机弹出一个
set1 = {'a','b','c',4}
set1.pop()
print(set1)
```

```python
# remove() 删除集合中的一个值，如果找不到就报错
set1 = {5,1,2,3,4,'b','u'}
set1.remove(5)
print(set1)
set1.remove('5')
```

```python
# discard() 删除集合中的一个值，如果找不到就什么也不做
set1 = {5,1,2,3,4,'b','u'}
set1.discard(3)
print(set1)
set1.discard(3)
print(set1)
print(set1.discard(2))
```

```python
# difference()差集  集合A中的元素减去集合B中相同的元素后，剩下的元素
set1 = {1,2,3,4}
set2 = {1,8,3,5,6,7}
print(set1)
print(set2)
print(set2.difference(set1))
print('='*20)

print(set2 - set1)#也可以直接用减号
print('='*20)

# difference_update() 区别就是前者是返回一个新的集合，后者是覆盖原来的集合，#返回值位None
set2.difference_update(set1)
print(set2)
```
